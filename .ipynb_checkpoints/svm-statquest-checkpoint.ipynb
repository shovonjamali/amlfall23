## Missing Data Part 1: Identifying Missing Data
Unfortunately, the biggest part of any data analysis project is making sure that the data are correclty formatted and fixing it when it is not. The first part of this process is identifying and dealing with the missing data.

**Missing Data** is simply a blank space, or a surrogate value like **NA**, that indicates that we failed to collect data for one of the features. If we forgot to ask someone's age, or to forgot to write it down, then we would have a blank space in the dataset for that person's **age**.

There are two main ways to deal with the missing data:

1. We can remove the rows that contain the missing data from the dataset. This is relatively easy to do but wastes all of the other values we collected. How a big of waste this is depends on how important this missing value is for classification. For example, if we are missing a value for **age** and **age** is not useful for classifying if people have heart disease or not, then it would be a shame to throw out all of someone's data just because we do not have their **age**. 
2. We can **impute** the values that are missing. In this context **impute** is just a fancy way of saying "we can make an educated guess about what the value should be". Continuing our example where we are missing a value for **age**, instead of throwing out the entire row of the data, we can fill the missing value with the average age or the median age, or use some other, more sophisticated approach, to guess at an appropriate value.

In this section, we'll focus on identifying missing values in the dataset.

## Missing Data Part 2: Dealing With Missing Data

Since scikit-learn's support vector machine do not support dataset with missing values, we need to figure out what to do with the 0s in the dataset. We can either delete these customers from the training dataset, or impute values for the missing data. First, let's see how many rows contain missing values. 

```python
len(df.loc[(df['EDUCATION'] == 0) | (df['MARRIAGE'] == 0)])
```

**NOTE:** The breakdown of the above code is as follows:

'**df['EDUCATION'] == 0**' creates a Boolean Series. It checks whether the values in the 'EDUCATION' column of the DataFrame df are equal to 0. This operation will return a Series of Boolean values, where each element corresponds to a row in the DataFrame and indicates whether the 'EDUCATION' column for that row has the value 0.

```python
df['EDUCATION'] == 0
```

```python
df.loc[(df['EDUCATION'] == 0)]
```

```python
df.loc[(df['MARRIAGE'] == 0)]
```

```python
df.loc[(df['EDUCATION'] == 0) | (df['MARRIAGE'] == 0)]
```

So, only 68 rows have missing values. Now, let's count the total number of rows in the dataset:

```python
len(df)
```

So, 68 of the 30000 rows, or less than 1% contain missing values. Since that still leaves us with more data that we need for a **Support Vector Machine**, we will remove the rows with missing values, rather than impute thier values. We do this by selecting all the rows that do not contain **0** in either the **EDUCATION** or **MARRIAGE** columns:

```python
df_no_missing = df.loc[(df['EDUCATION'] != 0) & (df['MARRIAGE'] != 0)]
```

Since df_no_missing has **68** fewer rows than the original df, it should have **29,932** rows.

It works! Still, we can also make sure **EDUCATION** no longer contains **0s** by printing it's unique values:

```python
df_no_missing['EDUCATION'].unique()
```

And we can also do the same for **MARRIAGE**:

```python
df_no_missing['MARRIAGE'].unique()
```

**BAM**! We have verified that df_no_missing does not contain any missing values.

## Downsample the Data

**Support Vector Machines** are great with small datasets, but not awesome with large ones, and this dataset, while not huge, is big enough to take a long time to optimize with **Cross Validation**. So we'll downsample both categories, customers who did and did not default, to 1000 each.

To make sure we get **1000** of each category, we start by splitting the data into two **dataframes**, one for people that did not default, one for people that did.

```python
df_no_default = df_no_missing[df_no_missing['DEFAULT'] == 0]
df_default = df_no_missing[df_no_missing['DEFAULT'] == 1]
```

```python
df_no_default
```

```python
df_default
```

```python
df_no_default_downsampled = resample(df_no_default,
                                    replace=False,
                                    n_samples=1000,
                                    random_state=42)
len(df_no_default_downsampled)
```

Now, downsample the dateset that defaulted:

```python
df_default_downsampled = resample(df_default,
                                 replace=False,
                                 n_samples=1000,
                                 random_state=42)
len(df_default_downsampled)
```

Now, let's merge the two downsampled datasets into a single **dataframe** and print out the total number of samples to make sure everything is hunky dory.

```python
df_downsample = pd.concat([df_no_default_downsampled, df_default_downsampled])
len(df_downsample)
```

## Format Data Part 1: Split the Data into Dependent and Independent Variables

In the code below we are using the copy() function to copy the data by value. By default, panda uses copy by reference. Using copy(), ensures that original data df_downsample is not modified when we modify X or y.

```python
X = df_downsample.drop('DEFAULT', axis=1).copy()
X.head()
```

```python
y = df_downsample['DEFAULT'].copy()
y.head()
```

## Format the Data Part 2: One-hot Encoding

The dataset contains both numerical data as well as categorical data like **Marriage**. scikit-learn's **Support Vector Machine** natively supports continuous data rather than categorical data. So we need to transform some of the columns value so that it can be supported. This process is known as one hot encoding. There are many ways to do **One-hot Endocing** in python. Two of the most popular methods are ColumnTransformer() from **scikit-learn** and get_dummies() from **pandas**, and both methods have pros and cons. We will be using get_dummies as it's easier to understand how one-hot encoding works as it keeps the column names.

First, before we commit to converting columns with **One-hot Encoding**, let's just see what happens when we convert **MARRIAGE** without saving the results. This will easy to see how get_dummies() work.

```python
pd.get_dummies(X, columns=['MARRIAGE']).head()
```

```python
X_encoded = pd.get_dummies(X, columns=['SEX',
                                      'EDUCATION',
                                      'MARRIAGE',
                                      'PAY_0',
                                      'PAY_2',
                                      'PAY_3',
                                      'PAY_4',
                                      'PAY_5',
                                      'PAY_6'])
X_encoded.head()
```

## Format the Data Part 3: Centering and Scaling

The **Radical Basis Function (RBF)** that we are using with our **Support Vector Machine** assumes that the data are centered and scaled. In other words, each columns should have a mean value = 0 and standard deviation = 1. S we need to do this to both the training and testing datasets.

**NOTE**: We split the data into training and testing datasets and then scale them separately to avoid **Data Leakage**. **Data Leakage** occurs when infomation about the training dataset corrupts or influence the testing dataset.

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42)
X_train_scaled = scale(X_train)
X_test_scaled = scale(X_test)
```

## Build a Preliminary Support Vector Machine

```python
clf_svm = SVC(random_state=42) # Creating a shell for Support Vector Classifier
clf_svm.fit(X_train_scaled, y_train)
```

Ok, we have built a **Support Vector Machine** for classification. Let's see how it performs on the **Testing Dataset** and draw a **Confusion Matrix**.

```python
predictions = clf_svm.predict(X_test_scaled)
cm = confusion_matrix(y_test, predictions, labels=None) # clf_svm.classes_
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=['Did not default', 'Defaulted']) # ['Did not default', 'Defaulted'], clf_svm.classes_
disp.plot()
plt.show()
```

In the confusion matrix, we see that of the **257** people that did not default, **201 (79%)** were correctly classified. And of the **243** people that defaulted, **148 (61%)** were correctly classfied. So the **Support Vector Machine** was not awesome. So, let's try to improve prediction using **Cross Validation** to optimzie the parameters.

## Optimize Parameters with Cross Validation and GridSearchCV()

Optimizing a **Support Vector Machine** is all about finding the best value for **gamma**, and, potentially, the regularization parameter **C**. So, let's see if we can find better parameters values using cross validation in hope that we can improve the accurarcy with the **Testing Dataset**.

Since we have two parameters to optimize, we will use the GridSearchCV(). We specify a bunch of potential values for **gamma** and **C**, and GridSearchCV() tests all possible combinations of the parameters for us.

```python
param_grid = [
    {'C': [0.5, 1, 10, 100], # NOTE: Values for C must be > 0
     'gamma': ['scale', 1, 0.1, 0.01, 0.001, 0.0001],
     'kernel': ['rbf']}
]

## NOTE: We are inlcuding C=1 and gamma='scale'
## as possible choices since they are the default values.

optimal_params = GridSearchCV(
    SVC(),
    param_grid,
    cv=5,
    scoring='accuracy', ## (accuracy is default scoring)
    ##scoring='balanced_accuracy',
    ##scoring='f1',
    ##scoring='f1_micro',
    ##scoring='f1_macro',
    ##scoring='f1_weighted',
    ##scoring='roc_auc',
    verbose=0 # NOTE: If you want to see what Grid Search is doing, set verbose=2
)
optimal_params.fit(X_train_scaled, y_train)
print(optimal_params.best_params_)
```

And we see that the ideal value for C is **100**, which means we will use regularization, and the ideal value for **gamma** is **0.001**.

## Building, Evaluating, Drawing, and Interpreting the Final Support Vector Machine

```python
clf_svm = SVC(random_state=42, C=100, gamma=0.001)
clf_svm.fit(X_train_scaled, y_train)
```

Now, let's plot another confusion matrix to see if the optimized support vector machine does better. 

```python
predictions = clf_svm.predict(X_test_scaled)
cm = confusion_matrix(y_test, predictions, labels=None) # clf_svm.classes_
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=['Did not default', 'Defaulted']) # ['Did not default', 'Defaulted'], clf_svm.classes_
disp.plot()
plt.show()
```
